```python
#Python program to scrape website  
#and save quotes from website 
import requests 
from bs4 import BeautifulSoup 
import csv 
```

```python
URL = "http://www.values.com/inspirational-quotes"
r = requests.get(URL)
```

```python
soup = BeautifulSoup(r.content, 'html5lib') 
# html5lib : Specifying the HTML parser we want to use.
```

```python
# prettify : It is used to display the HTML in the nested structure.
print(soup.prettify())
```

```python
print(soup.title.get_text())
```

```python
quotes=[]  # a list to store quotes 
   
table = soup.find('div', attrs = {'id':'all_quotes'}) 
```

```python
print(table)
```

```python
# .findAll() method : It will return a list of all the elements in the tag.

for row in table.findAll('div', 
                         attrs = {'class':'col-6 col-lg-4 text-center margin-30px-bottom sm-margin-30px-top'}): 
    print(row)
    quote = {} 
    quote['theme'] = row.h5.text 
    quote['url'] = row.a['href'] 
    quote['img'] = row.img['src'] 
    quote['lines'] = row.img['alt'].split(" #")[0] 
    quote['author'] = row.img['alt'].split(" #")[1] 
    quotes.append(quote)
```

```python
print(quotes)
```

```python
import pandas as pd
print(pd.DataFrame(quotes).head())
```

```python
filename = 'inspirational_quotes.csv'

with open(filename, 'w', newline='') as f: 
    w = csv.DictWriter(f,['theme','url','img','lines','author']) 
    w.writeheader() 
    for quote in quotes: 
        w.writerow(quote) 
```
